## puzzle

The spreadsheet consists of rows of apparently-random numbers. To make sure the recovery process is on the right track, they need you to calculate the spreadsheet's checksum. For each row, determine the difference between the largest value and the smallest value; the checksum is the sum of all of these differences.

For example, given the following spreadsheet:

5 1 9 5
7 5 3
2 4 6 8
The first row's largest and smallest values are 9 and 1, and their difference is 8.
The second row's largest and smallest values are 7 and 3, and their difference is 4.
The third row's difference is 6.
In this example, the spreadsheet's checksum would be 8 + 4 + 6 = 18.

What is the checksum for the spreadsheet in your puzzle input?

````python
checksum = 0
for line in puzzle.splitlines():
    firstitem = True
    for item in line.split(sep='\t'):
        if firstitem is True:
            # intiialize the line min and max
            lmin = int(item)
            lmax = int(item)
            firstitem = False
        else:
            if int(item) < lmin:
                # lesser min found, replace line min
                lmin = int(item)
            if int(item) > lmax:
                # greater max found, replace the line max
                lmax = int(item)
    # generate the diff and add that to the checksum
    checksum = checksum + lmax - lmin
checksum
````
